```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
amostra = []

for n in range(1, 1000):
    
    vetor_normal = np.random.normal(0, 1, size=(n, 1000))
    maximo = np.max(vetor_normal, axis=0)
    media = np.mean(maximo)

    amostra.append(media)
```

```python
x = np.arange(1, 1000)
```

```python
plt.plot(amostra)
plt.plot((2*np.log(x))**.5)
```
