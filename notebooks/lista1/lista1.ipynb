# Ex 4

```python
def f(x):
    e = np.exp(1)
    return x*(1-x)*e**x/(3-e)
```

```python
def g(x):
    return 1
```

```python
e = np.exp(1)

x_max = (5**.5 / 2 - .5)

c = ((5**.5 - 2) * e**(-.5 + (5**.5)/2)) / (3-e)
```

```python
x = np.linspace(0,1,100)
```

```python
plt.plot(x,f(x))
plt.axvline(x_max)
plt.axhline(c)
```

```python
def amostra_de_f(c):

    u = 1
    quociente = 0

    contador_rejeicoes = 0
    
    while u > quociente:
        u = np.random.uniform()
        y = np.random.uniform() # sabemos amostrar disso!!!
    
        quociente = f(y) / (g(y) * c)

        contador_rejeicoes = contador_rejeicoes + 1

    return y, contador_rejeicoes
        
```

```python
tamanho_amostra = 10000

amostra = []
rejeicoes = []

for i in range(tamanho_amostra):
    y_de_f, n_rejeicoes = amostra_de_f(c)

    amostra.append(y_de_f)
    rejeicoes.append(n_rejeicoes)

    
```

```python
plt.hist(amostra, bins=50, density =True)
plt.plot(x, f(x))
```

## Ex 2

```python
def f(x):
    return np.exp(-2**.5 * abs(x))/ (2**.5)
```

```python
x = np.linspace(-5,5,200)
```

```python
plt.plot(x,f(x))
```

```python
def F(x):
    if x < 0:
        return np.exp(2**.5 * x) / 2
    else:
        return 1 - np.exp(-2**.5 * x) / 2

F = np.vectorize(F)
```

```python
plt.plot(x,F(x))
```

```python
def F_inv(y):
    if y <= .5:
        return np.log(2*y) / 2**.5
    else:
        return - np.log(2 - 2*y) / 2**.5

F_inv = np.vectorize(F_inv)
```

```python
plt.plot(x,F_inv(x))
```

```python
def amostra_de_F():
    u = np.random.uniform()

    return F_inv(u)
```

```python
tamanho_amostra = 10000

amostra = []

for i in range(tamanho_amostra):
    y_de_f = amostra_de_F()

    amostra.append(y_de_f)
```

```python
plt.hist(amostra, bins=50, density =True)
plt.plot(x, f(x))
```
