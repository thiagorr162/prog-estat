```python
import numpy as np
import matplotlib.pyplot as plt

p = .7

def sample_from_bias(p):

    quot = 0
    u = 1

    while u > quot:
        y = np.random.choice(2, p=[p, 1-p]) # g(y)
        
        c = max([.5/p, .5/(1-p)]) # max(f(y)/g(y))
        
        if y==0:
            quot = .5/(c*p)
        else:
            quot = .5/(c*(1-p))
    
        u = np.random.uniform()

    return y

amostra = []


for i in range(10000):
    amostra.append(sample_from_bias(p))

plt.hist(amostra)
```
