## Aproximação de Distribuição Normal a Partir de Variáveis Uniformes

Este código demonstra uma técnica para gerar dados com distribuição normal usando variáveis aleatórias uniformemente distribuídas. A ideia central é baseada no Teorema Central do Limite, que afirma que a soma de variáveis aleatórias independentes e identicamente distribuídas, como as variáveis uniformes, tende a uma distribuição normal conforme o número de variáveis aumenta.

### Explicação do Método

1. Geramos múltiplos conjuntos de variáveis aleatórias uniformes.
2. Somamos as variáveis em cada conjunto e aplicamos uma normalização para que a distribuição das somas se aproxime de uma normal padrão.
3. Finalmente, ajustamos a média e o desvio padrão dos valores para obter a distribuição normal desejada.

A visualização compara o histograma dos dados gerados com a curva de uma distribuição normal padrão esperada, demonstrando a eficácia do método.


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def generate_normal_via_uniform(num_samples=1000, n=1000):
    # Generate normal distribution approximation using uniform variables
    uniform_samples = np.random.uniform(0, 1, (num_samples, n))
    z_scores = np.sqrt(12 / n) * (uniform_samples.sum(axis=1) - n / 2)
    return z_scores

# Generate and plot data
generated_data = generate_normal_via_uniform(num_samples=10000)
plt.hist(generated_data, bins=50, density=True, alpha=0.6, color="blue", label="Simulado")

# Plot expected normal distribution for comparison
x = np.linspace(min(generated_data), max(generated_data), 1000)
plt.plot(x, stats.norm.pdf(x, 0, 1), 'r--', label="Densidade Real")

# Labels and legend
plt.xlabel("Eixo x")
plt.ylabel("Densidade")
plt.title("Gerado vs Real")
plt.legend()
plt.show()
```
