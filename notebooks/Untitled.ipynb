```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

amostra = []

size = 1000000

for i in range(size):

    u = np.random.uniform()
    u = np.ceil(6*u)
    amostra.append(u)

amostra = np.array(amostra)
```

```python
plt.hist(amostra)
```

```python
for i in range(1,7):

    print(sum(amostra == 5) / size)
```
